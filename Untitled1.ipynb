```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_palette("Set2")

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(r"C:\Users\BIT\OneDrive\jobmarketanalytics\data\indian_tech_jobs_2026.csv")

df.head()
```

```python
print("Rows :",df.shape[0])
print("Columns :",df.shape[1])

df.info()
```

```python
missing=df.isnull().sum().sort_values(ascending=False)

missing[missing>0]
```

```python
print("Duplicate Rows :",df.duplicated().sum())

df=df.drop_duplicates()

print("Shape after removing duplicates :",df.shape)
```

```python
df.describe(include='all')
```

```python
print("Total Jobs :",len(df))
print("Companies :",df.company_name.nunique())
print("Cities :",df.primary_city.nunique())
print("Average Company Rating :",round(df.company_rating.mean(),2))
print("Remote Jobs :",len(df[df.work_mode=="Remote"]))
```

```python
plt.figure(figsize=(12,6))

top_company=df.company_name.value_counts().head(10)

sns.barplot(x=top_company.values,
            y=top_company.index)

plt.title("Top Hiring Companies")

plt.xlabel("Job Postings")

plt.ylabel("Company")

plt.show()
```

```python
plt.figure(figsize=(10,6))

top_city=df.primary_city.value_counts().head(10)

sns.barplot(x=top_city.values,
            y=top_city.index)

plt.title("Top Hiring Cities")

plt.xlabel("Job Count")

plt.show()
```

```python
plt.figure(figsize=(10,5))

role=df.role_category.value_counts()

sns.barplot(x=role.index,
            y=role.values)

plt.xticks(rotation=30)

plt.title("Most In-Demand Roles")

plt.show()
```

```python
plt.figure(figsize=(6,6))

df.work_mode.value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")

plt.title("Work Mode Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df.company_rating,
             bins=20)

plt.title("Company Rating Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,
              x="experience_tier",
              order=df.experience_tier.value_counts().index)

plt.xticks(rotation=25)

plt.title("Jobs by Experience Level")

plt.show()
```

```python
# Filter jobs with disclosed salary
salary = df[df['salary_max_lpa'] > 0]

salary_role = (
    salary.groupby('role_category')['salary_max_lpa']
          .mean()
          .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))

sns.barplot(
    x=salary_role.index,
    y=salary_role.values
)

plt.xticks(rotation=30)
plt.xlabel("Role Category")
plt.ylabel("Average Maximum Salary (LPA)")
plt.title("Highest Paying Roles")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))

skill=df.skill_domain.value_counts()

sns.barplot(x=skill.index,
            y=skill.values)

plt.xticks(rotation=30)

plt.title("Top Skill Domains")

plt.show()
```

```python
salary = df[df['salary_tier'] != 'Undisclosed']

order = [
    'Entry (<3 LPA)',
    'Junior (3-6 LPA)',
    'Mid (6-10 LPA)',
    'Senior (10-20 LPA)',
    'Leadership (20+ LPA)'
]

plt.figure(figsize=(9,5))

sns.countplot(
    data=salary,
    x='salary_tier',
    order=order
)

plt.title("Salary Tier Distribution (Disclosed Salaries)")
plt.xlabel("Salary Tier")
plt.ylabel("Number of Jobs")

plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

```python
numeric=df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))

sns.heatmap(numeric.corr(),
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```
